## <span style='color: orange'>1. a) Les 10 pays ayant le plus haut ratio disponibilité alimentaire/habitant en termes de protéines (en kg) par habitant<a name='M111'></a>

```python
import sqlite3
```

```python
connexion = sqlite3.connect("baseP3.db")
curseur = connexion.cursor()
```

```python
curseur.execute('''
SELECT
    pays,
    ROUND(SUM(dispo_prot)/1000*365,2) AS dispo_prot_kg_h
FROM dispo_alim
WHERE année = 2013
GROUP BY pays
ORDER BY dispo_prot_kg_h DESC
LIMIT 10;
''')
for resultat in curseur:
    print(resultat)
```

## <span style='color: orange'>1. b) Les 10 pays ayant le plus haut ratio disponibilité alimentaire/habitant en termes de kcal par habitant<a name='M112'></a>

```python
curseur.execute('''
SELECT
    pays,
    SUM(dispo_alim_kcal_p_j)*365 AS dispo_alim_kcal_h
FROM dispo_alim
WHERE année = 2013
GROUP BY pays
ORDER BY dispo_alim_kcal_h DESC
LIMIT 10;
''')
for resultat in curseur:
    print(resultat)
```

## <span style='color: orange'>2. Pour chaque année disponible, les 10 pays ayant le plus faible ratio disponibilité alimentaire/habitant en termes de protéines (en kg) par habitant<a name='M12'></a>

```python
curseur.execute('''
SELECT *
FROM (
SELECT
    année,
    pays,
    ROUND(SUM(dispo_prot)/1000*365,2) AS dispo_prot_kg_h
FROM dispo_alim
WHERE année = 2012
GROUP BY
    pays,
    année
ORDER BY
    dispo_prot_kg_h
LIMIT 10)
UNION
SELECT *
FROM (
SELECT
    année,
    pays,
    ROUND(SUM(dispo_prot)/1000*365,2) AS dispo_prot_kg_h
FROM dispo_alim
WHERE année = 2013
GROUP BY
    pays,
    année
ORDER BY
    dispo_prot_kg_h
LIMIT 10);
''')
for resultat in curseur:
    print(resultat)
```

## <span style='color: orange'>3. La quantité totale (en kg) de produits perdus par pays et par année<a name='M13'></a>

```python
curseur.execute('''
SELECT 
      année,
      pays,
      SUM(pertes)*1000000 AS pertes_kg
FROM equilibre_prod
GROUP BY
      pays,
      année;
''')
for resultat in curseur:
    print(resultat)
```

## <span style='color: orange'>4. Les 10 pays pour lesquels la proportion de personnes sous-alimentées est la plus forte<a name='M14'></a>

```python
curseur.execute('''
SELECT
    sn.pays,
    CASE
        WHEN sn.nb_personnes != 100000 THEN ROUND(
                                                  CAST(sn.nb_personnes AS REAL) / p.population * 100     /*Le type de résultat d'une opération est déduit du type de chacun des termes, le choix étant fait sur le type le plus large*/
                                                  , 1)
    END proportion
FROM sous_nutrition sn
JOIN population p
ON sn.code_pays = p.code_pays AND sn.année = p.année
WHERE sn.année = 2013
ORDER BY proportion DESC
LIMIT 10;
''')
for resultat in curseur:
    print(resultat)
```

## <span style='color: orange'>5. Les 10 produits pour lesquels le ratio Autres utilisations/Disponibilité intérieure est le plus élevé<a name='M15'></a>

```python
curseur.execute('''
SELECT
    produit,
    ROUND(
          CAST(SUM(autres_utilisations) AS REAL) / SUM(dispo_int) * 100
          , 2) AS Ratio
FROM equilibre_prod
GROUP BY produit
ORDER BY Ratio DESC
LIMIT 10;
''')
for resultat in curseur:
    print(resultat)
```
