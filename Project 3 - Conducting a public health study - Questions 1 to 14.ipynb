# Question 4:<a name='M4'></a>
## <span style='color:orange'>1. Calcul pour chaque pays et chaque produit des ratios Energie/Poids<a name='M41'></a>

```python
calo1 = bil(['Disponibilité alimentaire (Kcal/personne/jour)'])
calo1['Valeur'] = calo1['Valeur'].mul(calo1['Population'], axis = 'rows')
calo1 = pd.pivot_table(calo1, "Valeur", ['Pays','Population','Produit'], "Élément").fillna(0)
calo1.head()
```

```python
calo2 = bil(['Nourriture'])
calo2 = pd.pivot_table(calo2, "Valeur", ['Pays', 'Population', 'Produit'], "Élément").fillna(0)
calo = calo1.join(calo2)
calo["Disponibilité alimentaire (Kcal/personne/jour)"] *= 365
calo['Nourriture'] *= 1000000
calo.columns = ["Disponibilité alimentaire (kcal)", 'Nourriture (millions de kg)']
calo = calo.apply(np.int64)
calo = calo[(calo[['Disponibilité alimentaire (kcal)', 'Nourriture (millions de kg)']] > 0).all(axis = 1)] # supprime les zéros des colonnes (utile pour le futur calcul du ratio)
calo['Ratio "énergie/poids" (kcal/kg)'] = calo['Disponibilité alimentaire (kcal)'].div(calo['Nourriture (millions de kg)'], fill_value = 0)
calo = calo.reset_index()
calo = calo.drop(['Disponibilité alimentaire (kcal)', 'Nourriture (millions de kg)', 'Population'],1)
calo['Ratio "énergie/poids" (kcal/kg)'] = calo['Ratio "énergie/poids" (kcal/kg)'].round(decimals = 0)
calo
```

```python
calo_oeuf.describe().apply(np.int64)
```

## <span style='color:orange'>2) Calcul pour chaque pays et chaque produit du pourcentage de protéines<a name='M42'></a>

```python
prot = bil(['Disponibilité de protéines en quantité (g/personne/jour)', 'Disponibilité alimentaire en quantité (kg/personne/an)'])
prot = prot.drop(['Population'], 1)
prot = pd.pivot_table(prot, 'Valeur', ['Pays', 'Produit'], 'Élément').fillna(0)
prot["Disponibilité de protéines en quantité (g/personne/jour)"] *= 365
prot["Disponibilité alimentaire en quantité (kg/personne/an)"] *= 1000
prot.columns = ["Disponibilité alimentaire en quantité (g/personne/an)","Disponibilité de protéines en quantité (g/personne/an)"]
prot = prot[(prot[['Disponibilité alimentaire en quantité (g/personne/an)', 'Disponibilité de protéines en quantité (g/personne/an)']] != 0).all(axis = 1)]  # supprime les zéros des colonnes (utile pour le futur calcul du ratio)
prot['Pourcentage de protéines'] = (prot['Disponibilité de protéines en quantité (g/personne/an)'].mul(100)).div(prot['Disponibilité alimentaire en quantité (g/personne/an)'], fill_value = 0)
prot = prot.reset_index()
prot = prot.drop(['Disponibilité alimentaire en quantité (g/personne/an)','Disponibilité de protéines en quantité (g/personne/an)'],1)
prot.head().round(decimals = 2)
```

```python
ratio1 = bil(['Autres Utilisations', 'Disponibilité intérieure'])
ratio1 = ratio1.drop(['Population'], 1)
ratio1 = ratio1.loc[ratio1['Produit'].isin(list(imports['Produit']))]
ratio1 = pd.pivot_table(ratio1, 'Valeur', ['Pays','Produit'], 'Élément').fillna(0)
ratio1['Ratio'] = (ratio1['Autres Utilisations'].mul(100)).div(ratio1['Disponibilité intérieure']).fillna(0).round(decimals = 2)
ratio1
```
